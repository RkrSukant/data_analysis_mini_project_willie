### Previous Purchases (Using KNN)

```python
# Getting features to use to identify nearest neighbour
features_for_knn = ['age', 'purchase_amount', 'previous_purchases']

# Scale the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df[features_for_knn])

# 3. Initialize and apply the KNN Imputer
# n_neighbors=5 is a standard starting point for this dataset size
imputer = KNNImputer(n_neighbors=5)
df_imputed_array = imputer.fit_transform(df_scaled)

# 4. Inverse transform to get the original scale back
df_final_values = scaler.inverse_transform(df_imputed_array)

# 5. Update the 'previous_purchases' column in your main dataframe
# We use round() because you cannot have a fraction of a purchase
df['previous_purchases'] = df_final_values[:, 2].round()
```

```python
df.isnull().sum()
```

### Purchase Amount (USD)

```python
df.groupby("item_purchased")['purchase_amount'].mean()
```

```python
# Filling null values of Purchase Amount with the mean of each Item Purchased
df['purchase_amount']=df['purchase_amount'].fillna(df.groupby('item_purchased')['purchase_amount'] .transform ('mean'))
```
